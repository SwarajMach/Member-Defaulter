# Logistic Model

```python
# split X and y into training and testing sets
from sklearn.cross_validation import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=0)
```

```python
# train a logistic regression model on the training set
from sklearn.linear_model import LogisticRegression

# instantiate model
logreg = LogisticRegression()

# fit model
model = logreg.fit(X1_train, y1_train)
```

```python
# make class predictions for the testing set
y1_pred_class = model.predict(X1_test)
```

# Decision Tree Model

```python
#Import Library
from sklearn import tree
from sklearn.metrics import roc_auc_score
```

```python
# split X and y into training and testing sets
from sklearn.cross_validation import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.3,random_state=0)
```

```python
# Create tree object 
decision = tree.DecisionTreeClassifier() 
# for classification, here you can change the algorithm as gini or entropy (information gain) by default it is gini  
# model = tree.DecisionTreeRegressor() for regression
decision
```

```python
# Train the model using the training sets and check score
decision.fit(X1_train, y1_train)
decision.score(X1_train, y1_train)
```

```python
#Predict Output
y_predict= decision.predict(X1_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y1_test, y_predict)
```

```python
#roc_auc_score(y,obj[:,1])
result = decision.predict_proba(X1_test)
result
```

```python
roc_auc_score(y1_test,result[:,1])
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

pd.DataFrame(
    confusion_matrix(y1_test, y_predict),
    columns=['Predicted 0', 'Predicted 1'],
    index=['True 0', 'True 1']
)
```
